```python
#import file
import pandas as pd
filepath = ("/content/kc_house_data.csv")
read_file = pd.read_csv(filepath)
df = pd.DataFrame(read_file)

print(df)


```

```python
rows, columns = df.shape
print("Number of rows:", rows)
print("Number of columns:", columns)
```

```python
#count duplicate rows
df.duplicated().sum()
```

```python
#count duplicate columns
duplicate_cols = df.columns[df.T.duplicated()].tolist()
print("Duplicate columns:", duplicate_cols)
```

```python
df.isnull().values.any()
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
print(df.head())
```

```python
df['bathrooms'] = df['bathrooms'].astype(int)
print(df.head())

df['floors'] = df['floors'].astype(int)
print(df.head())
```

```python
df.drop('view', axis=1, inplace=True)
df.drop('waterfront',axis=1,inplace=True)
```

```python
print(df.head())
```

```python
df[df['condition'] == 5].value_counts()
```

```python
df['price'] = df['price'].astype(int)
```

```python
data = df[df['zipcode'].astype(str).str.len() <= 6]
print(data)
```

```python
df['zipcode'].astype(str).str.len().value_counts()
```

```python
renovated_counts = df[df['yr_renovated'] != 0]['yr_renovated'].value_counts().sort_index()
print(renovated_counts)
```

```python
max_floor= df['floors'].max()
print(f"Maximum Floor :{max_floor}")
```

```python
print(df)
```

```python
house_after_2009 = df[df['yr_built'] >2009]
print(house_after_2009)

count_after_2009 = len(house_after_2009)
print(f"Number of houses built after 2009: {count_after_2009}")
```

```python
house_before_2009 = df[df['yr_built'] < 2009]
print(house_before_2009)

count_before_2009 = len(house_before_2009)
print(f"Number of houses built before 2009: {count_before_2009}")
```

```python
max_grade = df['grade'].max()
print(f"Maximum Grade: {max_grade}")
```

```python
house_grade_13 = df[df['grade'] == 13]
print(house_grade_13)

count_grade_13 = len(house_grade_13)
print(f"Number of houses with grade 13: {count_grade_13}")
```

```python
#max bedroom

max_bedroom = df['bedrooms'].max()
print(f"Maximum Bedroom: {max_bedroom}")

house_bedroom = df[df['bedrooms'] == 5]
print(house_bedroom)

house_bedrooms = len(house_bedroom)
print(f"Number of houses with bedroom 5: {house_bedrooms}")
```

```python
#fetch bedroom  33 data

bedroom_33 = df[df['bedrooms'] == 33]
print(bedroom_33)
```
